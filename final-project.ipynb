```python
# ðŸš€ Importing some libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Netflix dataset
df = pd.read_csv("data/netflix_titles.csv")

# Show the first 5 rows to confirm it loaded correctly
df.head()
```

```python
# Check the structure of the dataset
df.info()

# See basic statistics (mostly works for numeric columns)
df.describe()

# Look at column names
df.columns
```

# Handling Missing Values

```python
# Replace missing values in text columns with "Unknown"
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')
```

# Convert the date to a real datetime type

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

# Split the duration into numbers

```python
df['duration_int'] = df['duration'].str.extract('(\d+)').astype(float)
```
